```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a simple dataset
data = {
    "Patient": ["A", "B", "C", "D", "E"],
    "Anxiety_Score": [12, 18, 5, 23, 16],
    "Depression_Score": [20, 15, 10, 25, 12]
}

df = pd.DataFrame(data)

# Showcasing the dataset
print("Patient Data:\n", df)

# Plotting the data
df.plot(x="Patient", y=["Anxiety_Score", "Depression_Score"], kind="bar")
plt.title("Mental Health Scores by Patient")
plt.xlabel("Patient")
plt.ylabel("Score")
plt.show()
```
